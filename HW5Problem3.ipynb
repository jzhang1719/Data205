```python
sql="""
SELECT en.*
  FROM icij.PD_Officers o
     INNER JOIN icij.PD_Edges ed
       ON ed.node_1=o.node_id
     INNER JOIN icij.PD_Entities en
       on en.node_id=ed.node_2
 WHERE o.name LIKE 'Tillerson%'
"""
df1=cj.executeQuery(sql,'IPDSDB')
df1
```

Part c:

This first query gives many more officers than in the graph.

```python
sql="""
SELECT o.*
  FROM icij.PD_Officers o1
     INNER JOIN icij.PD_Edges ed
       ON ed.node_1=o1.node_id
     INNER JOIN icij.PD_Entities en
       on en.node_id=ed.node_2
     INNER JOIN icij.PD_Edges ed1
       ON ed1.node_2=en.node_id
     INNER JOIN icij.PD_Officers o
       ON o.node_id=ed1.node_1
 WHERE o1.name LIKE 'Tillerson%'
"""
df2=cj.executeQuery(sql,'IPDSDB')
df2
```

To get only the officers with a relation to Yemen, add a LIKE clause to the WHERE condition.

```python
sql="""
SELECT o.*
  FROM icij.PD_Officers o1
     INNER JOIN icij.PD_Edges ed
       ON ed.node_1=o1.node_id
     INNER JOIN icij.PD_Entities en
       on en.node_id=ed.node_2
     INNER JOIN icij.PD_Edges ed1
       ON ed1.node_2=en.node_id
     INNER JOIN icij.PD_Officers o
       ON o.node_id=ed1.node_1
 WHERE o1.name LIKE 'Tillerson%' AND o.name LIKE 'Yemen%'
"""
df2=cj.executeQuery(sql,'IPDSDB')
df2
```
