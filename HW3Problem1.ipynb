```python
import numpy as np
import wave
import matplotlib.pyplot as plt
import wavio
```

Read in data using wavio

```python
wav1=wavio.read('/home/idies/workspace/AS.171.205/data/nylon.wav')
wav2=wavio.read('/home/idies/workspace/AS.171.205/data/strat.wav')
print(wav1)
print(wav2)
```

Convert into usable data using numpy fft

```python
freq1=abs(np.fft.rfft(wav1.data))
freq2=abs(np.fft.rfft(wav2.data))
freqs1 = np.fft.fftfreq(len(freq1))
freqs2 = np.fft.fftfreq(len(freq2))
```

Graph for Nylon.wav

The graphs don't look great as it is hard to distinguish the lines. 

Graph for Wav data

```python
plt.xlabel(r'$\mathrm{Time}$', fontsize=14)
plt.ylabel(r'$\mathrm{Value}$', fontsize=14)
plt.plot(wav1.data)
```

Graph for Frequency Spectrum

```python
plt.title("Frequency Spectrum for Nylon.wav")
plt.xlabel(r'$\mathrm{Frequency}$', fontsize=14)
plt.ylabel(r'$\mathrm{Amplitude}$', fontsize=14)
plt.xscale('log')
plt.plot(freq1)
```

Find Fundamental Frequency for Nylon

```python
ffi1=np.argmax(np.abs(freq1))
ff1=freqs1[ffi1]
# 44100 is sampling rate
ff1=ff1*44100
print('Fundamental Frequency for Strat is', ff1, 'Hz')
```
