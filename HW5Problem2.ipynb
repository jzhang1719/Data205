For every two points, find rho and theta values that represent the line that intersects them.

```python
#rho=[]
#theta=[]
#for i in range(0,1559):
#    for j in range(i+1,1560):
#        m = (df[1][i]-df[1][j])/(df[0][i]-df[0][j])
#        b = (df[0][i]*df[1][i] - df[0][j]*df[1][i])/(df[0][i]-df[0][j])
#        th = math.atan(-1 / m)
#        r = b * math.sin(th)
#        rho.append(r)
#        theta.append(th)

rho=[]
theta=[]
for i in range(0,1559):
    for j in range(i+1,1560):
        x1 = df[0][i]
        x2 = df[0][j]
        y1 = df[1][i]
        y2 = df[1][j]
        if y2 == y1:
            th = math.pi / 2
        else:
            th = math.atan((x1-x2)/(y2-y1))
        r = x1 * math.cos(th) + y1 * math.sin(th)
        rho.append(r)
        theta.append(th)
```

```python
# Rho and theta vectors
rho1 = np.array(rho)
theta1 = np.array(theta)

#rho1 = np.nan_to_num(rho1)
#theta1 = np.nan_to_num(theta1)
#zero = np.multiply(rho1,theta1)
#rho1 = rho1[np.nonzero(zero)]
#theta1 = theta1[np.nonzero(zero)]
#theta1.size
```
